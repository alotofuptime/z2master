# DRY!
> We found out how to filter our dataframe by the modules column. Clearly it works & clearly it is repetitve and we need to use this functionality time and time again in our code. Let's go ahead and create a function to solve this once and for all. We'll call our function `course_advisor` for now.

### Input
keyword: A string representing a programming langauge, topic or technology of interest

### Output
A new data frame filtered based on whether the given keyword exists in any of the modules columns

```python
def course_advisor(keyword):
    acros = ["ui", "ux", "api", "html", "css", "sql", "js"]
    try:
        if keyword.lower() in acros:
            keyword = keyword.upper()
        else:
            keyword = keyword.title()
        keyword_mask = ztm_api.modules.apply(lambda x: keyword in x)
        courses = ztm_api[keyword_mask]
        if courses.empty:
            return f"'{keyword}' doesn't match any search results. Try being more specific."
        return courses
    except AttributeError:
        return f"Invalid keyword: {keyword} is of type {type(keyword)}. Keyword must be str."
```

```python
course_advisor("api")
```
